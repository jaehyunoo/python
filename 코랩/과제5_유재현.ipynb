## **판다스의 DataFrame()함수로 데이터프레임 만들기**

```python
import pandas as pd # 판다스 함수 라이브러리 가져오기

# 판다스 데이터프레임을 딕셔너리 구조를 이용하여 P자동차 회사의 차종별 마력, 총중량, 연비 자료생성
df_p = pd.DataFrame({
    'name' : ['A', 'B', 'C', 'D', 'E', 'F', 'G'],
    'horse power' : [130 , 250, 190, 300, 210, 220, 170],
    'weight' : [1.9, 2.6, 2.2, 2.9, 2.4, 2.3, 2.2],
    'efficiency' : [16.3, 10.2, 11.1, 7.1, 12.1, 13.2, 14.2] })

print(df_p)
```

# 12.2-1) 이 데이터프레임의 name 열을 인덱스로 사용하고 싶다.
# set_index('name')을 데이터프레임에 적용하여
# 다음과 같은 결과가 나오게 만들어 보라.

```python
# 인덱스 변경 ( name열을 인덱스로 사용) => set_index('인덱스로 사용하고 싶은 열', inplace = True)
# inplace=True => 원본 데이터 프레임에 적용

df_p.set_index('name',inplace = True)

df_p
```

## **12.2-2) 가장 높은 마력을 가진 차종을 찾아, 다음처럼 중량과 연비를 확인해 보자.**

```python
# 'horse power'가 최대값인 행 출력
max_horse_power = df_p[df_p['horse power'] == df_p['horse power'].max()]
print(max_horse_power)

# df_p['horse power'] == df_p['horse power'].max() 각 원소는 'horse power' 열의 해당 원소가 최댓값과 같으면 True이고,  그렇지 않으면 False
```

# **12.2-3) 마력과 연비를 곱한 값이 가장 큰 차종을 다음과 같이 찾아 보라.**

```python
# 마력과 연비를 곱한 값을 열에 추가시킨다.
# 마력에 연비를 곱한값으로 지정
df_p['hp x mile'] = df_p['horse power'] * df_p['efficiency']

df_p
```

```python
# 추가시킨 뒤 마력과 연비를 곱한 값이 가장 큰차종을 찾는다
max_hp_mile = df_p[df_p['hp x mile'] == df_p['hp x mile'].max()]
print(max_hp_mile)

# df_p['hp x mile'] == df_p['hp x mile'].max() 각 원소는 'hp x mile' 열의 해당 원소가 최댓값과 같으면 True이고,  그렇지 않으면 False
```

## **12.3-1) Q 자동차 회사의 표를 가지고 판다스 데이터 프레임으로 만들어 출력해보라**

```python
# 판다스 데이터프레임을 딕셔너리 구조를 이용하여 Q자동차 회사의 차종별 마력, 총중량, 연비 자료생성
df_q = pd.DataFrame({
    'name' : ['A', 'B', 'C', 'D'],
    'horse power' : [120 , 220, 120, 200],
    'weight' : [1.9, 2.1, 1.5, 2.9],
    'efficiency' : [18.3, 19.2, 21.1, 17.3] })

# 인덱스 변경 ( name열을 인덱스로 사용) => set_index('인덱스로 사용하고 싶은 열', inplace = True)
# inplace=True => 원본 데이터 프레임에 적용
df_q.set_index('name',inplace = True)

df_q
```

## **12.3-2) Q 사의 데이터와  P사의 데이터를 합쳐 다음과 같은 하나의 데이터 프레임을 만들어라**

```python
# 문제에서의 표에는 hp x mile 항목이 없기때문에 위에 df_p에 추가해줬던 hp x mile 항목을
# del을 이용해 없애준다
del df_p['hp x mile']

# pandas.concat([df_p,df_q]) df_p,df_q 두 데이터프레임을 합치는 concat() 함수 사용하기.
new_df = pd.concat([df_p, df_q])

new_df
```

## **12.3-3) 각 차량의 제조사를 알 수 있게 다음과 같이 데이터프레임을 다시 만들어 보라**

```python
# P회사에 제조사명(com)을 추가해 P 입력
df_p['com'] = 'P'
# Q회사에 제조사명(com)을 추가해 Q입력
df_q['com'] = 'Q'
# com항목을 추가한뒤 다시 df_p와 df_q의 데이터프레임을 합친다.
new_df = pd.concat([df_p, df_q])

new_df
```

## **12.3-4)각 회사에서 생산하는 차량들의 마력 x 연비의 평균값을 구해 회사별로 출력해보라**

```python
# 두 데이터프레임을 합친 데이터프레임에 hp x mile 항목을 만들어 각 차량마다 마력과 연비의 곱한값을 구한다
new_df['hp x mile'] = (new_df['horse power'] * new_df['efficiency'])

# 제조사(com)기준으로 그룹화하여 제조사(com)별  마력 X 연비의 평균값을 구해 회사별로 출력
# as_index=False => 데이터프레임으로 출력 됨.
mean_by_gender = new_df.groupby(by=['com'], as_index=False)['hp x mile'].mean()

mean_by_gender
```
