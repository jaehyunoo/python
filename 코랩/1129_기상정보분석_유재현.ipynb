```python
# 판다스 관련 라이브러리 가져오기
import pandas as pd

# 데이터를 데이터프레임 형태로 읽어오기
df = pd.read_csv('부산기온.csv', encoding = 'cp949')
df.head() # 기본 위에서 5줄만 보여준다.
```

## 결측치 처리
### 데이터프레임의 개요를 출력하여
### 결측치가 있는지 간단히 확인.
### info()함수로 데이터 프레임 df의
### 열별 개수와 자료형 출력

```python
# 데이터프레임 개요 확인
df.info()

# 최저기온, 최저기온 시각이 364개로 1개 모자람
```

### 데이터의 일부가 빠져 결측치가 있을 경우 데이터 하나를 그 이전 행 값으로 채워 넣기.
### 결측치를 대체할 때는 신중히 판단해야 한다.
### 결측의 원인과 특성을 고려하여 데이터의 분포를 왜곡하지 않도록 주의해야 한다.

```python
# 결측치 채우기 - 이전 행의 값과 같게 채움.
# df 원본 남기기
df2 = df.fillna(method='ffill')

df2.info()
```

## 그래프 생성.

### 기온 열의 이름에는 특수문자가 있다.
### 사용하기 편하도록 rename()함수를 사용하여 이름을 변경해서 사용하자.

```python
df2.rename(columns={'평균기온(°C)': '평균기온'}, inplace=True)
df2.rename(columns={'최고기온(°C)': '최고기온'}, inplace=True)
df2.rename(columns={'최저기온(°C)': '최저기온'}, inplace=True)

df2.head(3)
```

## 그래프(차트) 생성.
### <br>
### x축 : 1~365

```python
# 맷플롯립 관련 라이브러리 가져오기
import matplotlib.pyplot as plt

# 글꼴 설정
plt.rc('font', family = 'NanumGothic')

# 마이너스 기호가 깨지지 않도록 설정.
plt.rcParams['axes.unicode_minus'] = False

# 그래프의 제목 설정.
plt.title('부산시 2023년도 기온 변화')

# 선그래프 그리기
# x축과 y축 설정
plt.plot(range(1,len(df2)+1), df2['최고기온'],label = '최고기온', c='r')
plt.plot(range(1,len(df2)+1), df2['최저기온'],label = '최저기온', c='b')
plt.plot(range(1,len(df2)+1), df2['평균기온'],label = '평균기온', c='g')

# x축과 y축 라벨 설정
plt.xlabel('일')
plt.ylabel('기온')

# 범례 표시
plt.legend()

# 그래프 화면에 표시
plt.show()

# 파일로 저장.
plt.savefig('2023부산기온.png')
```

## 3월 데이터만 추출하기
### 일시는 Object 자료형이다.
### 3월 데이터를 추출할 수 있도록 "연-월-일"의 날짜형 데이터로 변경하자.

```python
# 일시 데이터를 날짜형으로 변경
df2['일시'] = pd.to_datetime(df2['일시'], format='%Y-%m-%d')

df2['일시']
```

```python
# 일시 열을 인덱스로 지정
# 일시열은 각 행에서 교유하면서 순차로 증가하기 때문에 이 열을 인덱스로 지정하면 편리함.
df3 = df2.set_index('일시')

df3.head()
```

```python
# 3월 데이터만 추출
df_Mar = df3[pd.DatetimeIndex(df3.index).month == 3]

df_Mar
```

```python
# 3월 기온의 선 그래프 생성.
plt.title('부산시 2023년도 3월 기온 변화')

plt.plot(range(1,32), df_Mar['최고기온'], label = '최고기온', c = 'r')
plt.plot(range(1,32), df_Mar['평균기온'], label = '평균기온', c = 'y')
plt.plot(range(1,32), df_Mar['최저기온'], label = '최저기온', c = 'b')

plt.xlabel('3월')
plt.ylabel('기온')

plt.xlim(1,31)  # x축 범위
plt.ylim(-7,22) # y축 범위

plt.legend()

plt.show()
plt.savefig('부산3월기온.png')
```
